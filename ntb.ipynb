```python
# Testing filters
from PIL import Image
import enchant
import pytesseract

# pytesseract.pytesseract.tesseract_cmd = r'tesseract'

# open image
img = Image.open('xxx.jpg')

# make image black and white
img = img.convert('L')

# show image using display library with max 300 width
from IPython.display import display
display(img.resize((300, int(img.height * 300 / img.width))))

# ocr image
text = pytesseract.image_to_string(img)

# print text
print(text)

# reverse image colors
img = Image.eval(img, lambda x: 255 - x)

# show image using display library with max 300 width
display(img.resize((300, int(img.height * 300 / img.width))))

# ocr image
text = pytesseract.image_to_string(img)

# print text
print(text)

# now apply Otsu's thresholding
from PIL import ImageFilter
img = img.filter(ImageFilter.MedianFilter())
img = img.point(lambda x: 0 if x < 128 else 255, '1')

# show image using display library with max 300 width
display(img.resize((300, int(img.height * 300 / img.width))))

# ocr image
text = pytesseract.image_to_string(img)

# print text
print(text)
```

```python
import cv2
import easyocr

image = cv2.imread('xxx.jpg', 0)

# threshold
ret, image = cv2.threshold(image, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)

# show image using display library with max 300 width
from IPython.display import display
display(Image.fromarray(image).resize((300, int(image.shape[0] * 300 / image.shape[1]))))


reader = easyocr.Reader(['en'])
result = reader.readtext(image)

# print each line with its confidence and only print above 60% confidence
for line in result:
    # convert str probability to float
    probability = float(line[2])
    if probability > 0.4:
        print(line[1], probability)

```
