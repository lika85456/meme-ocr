```python
import cv2
from sys import path
import matplotlib.pyplot as plt

path.append("../src")

import filter
import reader

# load "college_meme.jpg"
image = cv2.imread("./cs_meme_text.png", cv2.IMREAD_GRAYSCALE)

text = " ".join(reader.EasyOCRReader().read(image))


# set size to 12x12
plt.figure(figsize=(12, 12))

# show image and text next to each other, wrap the text
fig, (ax1, ax2) = plt.subplots(1, 2)
fig.backgroundcolor = "white"

# set width to 400px

ax1.imshow(image, cmap="gray")
ax1.axis("off")
textAx = ax2.text(-0.3, 0.2, text, wrap=True, fontsize=12)
textAx._get_wrap_line_width = lambda : 300.
ax2.axis("off")



plt.show()
```
